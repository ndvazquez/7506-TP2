```python
#cuidado hacer un left join, en estas features, porque creo que hay usuarios que no tienen viewed product.
import pandas as pd
df = pd.read_csv("../data/events.csv",low_memory=False)
pd.set_option('display.max_columns', 500)
df.head(5)
```

```python
viewedProductdf = df[df["event"]=="viewed product"].dropna(axis=1,how='all')
```

```python
viewedProductColordf = viewedProductdf.pivot_table(index='person', columns='color', values='timestamp', aggfunc='count').fillna(0)
```

```python
viewedProductColordf = viewedProductColordf.add_prefix('cant_col_vwd_prdct_')
```

```python
viewedProductColordf.head(5)
```

```python
viewedProductColordf.to_csv("../data/feature_cantidad_de_colores_vistos_vwd_product.csv", sep=',')
```

```python
checkOutdf = df[df["event"]=="checkout"].dropna(axis=1,how='all')
checkOutdfColorndf = checkOutdf.pivot_table(index='person', columns='color', values='timestamp', aggfunc='count').fillna(0).add_prefix('cant_col_checkout_')
```

```python
checkOutdfColorndf.head(5)
```

```python
checkOutdfColorndf.to_csv("../data/feature_cantidad_de_colores_vistos_checkout.csv", sep=',')
```
