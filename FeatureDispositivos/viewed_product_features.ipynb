```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('../data/events.csv', low_memory=False)
df['timestamp'] = pd.to_datetime(df['timestamp'])
df.head()
```

### Busco std de cantidad distinta de modelos vistos por usuario

```python
df_features = df.loc[df['event']=='viewed product', :]
agrupado = df_features.groupby(['person', 'model']).agg({'timestamp':'count'})
```

```python
agrupado.reset_index(inplace=True)
agrupado.columns = ['person', 'model', 'times viewed']
```

```python
agrupado = agrupado.groupby('person').agg({'times viewed':['mean', 'std', 'sum']})
agrupado.replace(to_replace=np.nan, value=0, inplace=True)
```

```python
agrupado.to_csv('../data/viewed_product_stats.csv')
```

### Busco los top k sku para los usuarios

```python
df_features = df.loc[df['sku']!=np.nan, :]
df_features.head()
```

```python
skus = df_features.loc[(df_features['event']=='viewed product') & (df_features['sku']!=np.nan) , ['sku', 'person']]
```

```python
skus_grouped = skus.groupby('person').agg({'sku':['mean', 'std']})
```

```python
skus_grouped.columns = ['sku mean', 'sku std']
```

```python
skus_grouped.replace(to_replace=np.nan, value=0, inplace=True)
```
