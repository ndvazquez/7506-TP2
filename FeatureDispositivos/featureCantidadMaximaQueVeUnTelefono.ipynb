```python
import pandas as pd
df = pd.read_csv('../data/events.csv', low_memory=False)
```

```python
df.head()
```

```python
df['timestamp'] =pd.to_datetime(df.timestamp)
df.sort_values(by=['person','timestamp'],inplace=True)
```

```python
vwdf = df[df["event"]=="viewed product"]
```

```python
vwdf
```

```python
vwdf= vwdf.loc[:,["timestamp","person","model","color","condition","storage"]]
```

```python
vwdf
```

```python
test = vwdf.groupby(["person","model"]).agg({'timestamp':'count'})
```

```python
test.reset_index(inplace=True)
```

```python
test = test.groupby("person").agg({"timestamp":"max"})
```

```python
personas_sin_repetir = df.drop_duplicates(subset="person", keep='first')
personas_sin_repetir = personas_sin_repetir.loc[:,"person"]
personas_sin_repetir = personas_sin_repetir.to_frame()
personas_sin_repetir.set_index("person",inplace=True)
```

```python
test.rename(columns={'timestamp': 'cantidadMaximaQueVeUnMismoProducto'}, inplace=True)
```

```python
df_final = personas_sin_repetir.join(test,on="person",how="left")
df_final.fillna(0,inplace=True)
```

```python
df_final.to_csv("../data/featureCantidadMaximaPersonaVeUnProducto.csv", sep=',')
```
