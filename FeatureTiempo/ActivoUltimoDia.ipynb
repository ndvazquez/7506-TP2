```python
import pandas as pd
```

```python
df = pd.read_csv('../data/events.csv', low_memory=False)
df['timestamp'] = pd.to_datetime(df['timestamp'])
df.sort_values(['timestamp'], ascending=True,inplace=True)
```

```python
df
```

```python
df2 = df.loc[df['timestamp'] >= pd.to_datetime('2018-05-31'), :]
```

```python
personas_sin_repetirDia31 = df2.drop_duplicates(subset="person", keep='first')
personas_sin_repetirDia31 = personas_sin_repetirDia31.loc[:,"person"]
personas_sin_repetirDia31 = personas_sin_repetirDia31.to_frame()
personas_sin_repetirDia31.set_index("person",inplace=True)
```

```python
personas_sin_repetirDia31["RealizoActividadEl31"]=True
```

```python
df3 = df.loc[df['timestamp'] >= pd.to_datetime('2018-05-30'), :]
df3 = df3.loc[df['timestamp'] < pd.to_datetime('2018-05-31'), :]
```

```python
personas_sin_repetirDia30 = df3.drop_duplicates(subset="person", keep='first')
personas_sin_repetirDia30 = personas_sin_repetirDia30.loc[:,"person"]
personas_sin_repetirDia30 = personas_sin_repetirDia30.to_frame()
personas_sin_repetirDia30.set_index("person",inplace=True)
personas_sin_repetirDia30["RealizoActividadEl30"]=True
```

```python
df4= df.loc[df['timestamp'] >= pd.to_datetime('2018-05-31 23:00:00'), :]
```

```python
personas_sin_repetirDia31UltimaHora = df4.drop_duplicates(subset="person", keep='first')
personas_sin_repetirDia31UltimaHora = personas_sin_repetirDia31UltimaHora.loc[:,"person"]
personas_sin_repetirDia31UltimaHora = personas_sin_repetirDia31UltimaHora.to_frame()
personas_sin_repetirDia31UltimaHora.set_index("person",inplace=True)
personas_sin_repetirDia31UltimaHora["RealizoActividadEl31UltimaHora"]=True
```

```python
df5 = df.loc[df['timestamp'] >= pd.to_datetime('2018-05-29'), :]
df5 = df3.loc[df['timestamp'] < pd.to_datetime('2018-05-30'), :]
personas_sin_repetirDia29 = df5.drop_duplicates(subset="person", keep='first')
personas_sin_repetirDia29 = personas_sin_repetirDia29.loc[:,"person"]
personas_sin_repetirDia29 = personas_sin_repetirDia29.to_frame()
personas_sin_repetirDia29.set_index("person",inplace=True)
personas_sin_repetirDia29["RealizoActividadEl29"]=True
```

## PREPARO EL FEATURE

```python
personas_sin_repetir = df.drop_duplicates(subset="person", keep='first')
personas_sin_repetir = personas_sin_repetir.loc[:,"person"]
personas_sin_repetir = personas_sin_repetir.to_frame()
personas_sin_repetir.set_index("person",inplace=True)
```

```python
df_final = personas_sin_repetir.join(personas_sin_repetirDia30,on="person",how="left")
df_final = df_final.join(personas_sin_repetirDia31,on="person",how="left")
df_final = df_final.join(personas_sin_repetirDia31UltimaHora,on="person",how="left")
df_final = df_final.join(personas_sin_repetirDia29,on="person",how="left")
```

```python
df_final.fillna(False,inplace=True)
```

```python
df_final.to_csv("../data/featureUsuariosRegistraronActividadEl293031YUltimaHora.csv", sep=',')
```
