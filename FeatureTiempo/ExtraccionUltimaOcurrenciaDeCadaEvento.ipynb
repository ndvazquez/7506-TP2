```python
import pandas as pd
df = pd.read_csv('../data/events.csv', low_memory=False)
```

```python
user_features = df[['person']].drop_duplicates().copy()
user_features.set_index('person', inplace=True)
```

```python
def ultima_ocurrencia(evento):
    df_resultado = df.loc[df['event'] == evento].groupby("person").agg({"timestamp":"max"})
    df_resultado['timestamp'] = df_resultado['timestamp'].str.replace("-","").str.replace(":","").str.replace(" ","")
    df_resultado['timestamp'] = pd.to_numeric(df_resultado['timestamp'])
    df_resultado.rename(columns={'timestamp': 'ult_'+evento}, inplace=True)
    return df_resultado
```

```python
ult_checkouts = ultima_ocurrencia('checkout')
user_features = user_features.join(ult_checkouts).fillna(0).astype(int)

ult_conversions = ultima_ocurrencia('conversion')
user_features = user_features.join(ult_conversions).fillna(0).astype(int)

ult_viewed_product = ultima_ocurrencia('viewed product')
user_features = user_features.join(ult_viewed_product).fillna(0).astype(int)

ult_searched_products = ultima_ocurrencia('searched products')
user_features = user_features.join(ult_searched_products).fillna(0).astype(int)

ult_visited_site = ultima_ocurrencia('visited site')
user_features = user_features.join(ult_visited_site).fillna(0).astype(int)

ult_leads = ultima_ocurrencia('lead')
user_features = user_features.join(ult_leads).fillna(0).astype(int)

ult_brand_listing = ultima_ocurrencia('brand listing')
user_features = user_features.join(ult_brand_listing).fillna(0).astype(int)

ult_generic_listing = ultima_ocurrencia('generic listing')
user_features = user_features.join(ult_generic_listing).fillna(0).astype(int)

ult_ad_camp = ultima_ocurrencia('ad campaign hit')
user_features = user_features.join(ult_ad_camp).fillna(0).astype(int)

ult_search_engine = ultima_ocurrencia('search engine hit')
user_features = user_features.join(ult_search_engine).fillna(0).astype(int)
```
